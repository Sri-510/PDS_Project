```python
from datetime import datetime

def change_dt(x):
    
    # Defining the datetime format
    f = "%Y-%m-%d %H:%M:%S"
    
    dt = datetime.strptime(x, f)
    
    # Initializing the minute variable
    m=0
    
    if dt.minute < 10:
        m = 0
    elif dt.minute >= 11 and dt.minute < 20:
        m = 10
    elif dt.minute >= 20 and dt.minute < 30:
        m = 20
    elif dt.minute >= 30 and dt.minute < 40:
        m = 30
    elif dt.minute >= 40 and dt.minute < 50:
        m = 40
    else:
        m = 50
    
    return dt.replace(minute=m, second=0).strftime(f)
```

```python
df.loc[:,'TIME'] = df["TIME"].apply(change_dt)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

#This above line imports the seasonal_decompose function from the statsmodels.tsa.seasonal module. 
#The statsmodels library is commonly used for statistical modeling and analysis, and tsa stands for time series analysis.

result = seasonal_decompose(df['SPEED'],period = 1008)  
result.plot();

#STL decomposing time series into seasonal, trend, and residualÂ components
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

#Exponential Smoothing State Space Models (ETS):ETS models and captures exponential trends and seasonality in timeÂ seriesÂ data.

df['TESadd12'] = ExponentialSmoothing(df['SPEED'],trend='add',seasonal='add',seasonal_periods=1008).fit().fittedvalues
df.head()
```

```python
df[['SPEED','TESadd12']].plot(figsize=(12,6)).autoscale(axis='x',tight=True);
```

```python
df[['SPEED','TESadd12']].iloc[80000:80500].plot(figsize=(12,6)).autoscale(axis='x',tight=True);
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
fitted_model = ExponentialSmoothing(df['SPEED']['6 jun,2023':'21 Aug, 2023'],trend='add',seasonal='add',seasonal_periods=1008).fit()
```

# Evaluating Model against Test Set

```python
test_predictions = fitted_model.forecast(2500).rename('Forecast')
```
